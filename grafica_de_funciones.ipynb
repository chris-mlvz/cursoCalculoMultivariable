```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x = np.linspace(0, 2*np.pi, 200)
y = np.sin(x)
plt.plot(x,y)
```

```python
y = np.cos(x)
plt.plot(x,y)
```

```python
x = np.linspace(0, 2*np.pi, 3)
y = np.cos(x)
plt.plot(x, y)
```

```python
x = np.linspace(0, 2*np.pi, 200)
y = x * np.exp(-x**2)
plt.plot(x, y)
plt.grid()
```

```python
plt.plot(x, y, 'r*')
```

```python
plt.plot(x, y, 'b--')
```
