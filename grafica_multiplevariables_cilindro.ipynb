```python
import numpy as np 
import matplotlib.pyplot as plt 
import matplotlib.cm as cm
%matplotlib inline

```

```python
# Creamos el rango de valores para el radio y la altura.
r = np.linspace(-5,5,20)
h = np.linspace(-5,5,20)

# Calculamos las coordenadas de los puntos (r,h).
r,h = np.meshgrid(r,h)

# Ecuación del volumen del cilindro.
z = np.pi * r**2 * h

# Graficamos la función.
fig = plt.figure()
ax = plt.axes(projection='3d')
p = ax.plot_surface(r,h,z, cmap=cm.coolwarm)
fig.colorbar(p, ax=ax)
```
