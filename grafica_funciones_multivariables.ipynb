```python
import numpy as np 
import matplotlib.pyplot as plt 
from matplotlib import cm
```

```python
u = np.arange(-2, 2, 0.1)
v = np.arange(-2, 2, 0.1)

x, y = np.meshgrid(u, v)
z = np.exp(-x**2 - y**2)
```

```python
fig = plt.figure()
fig.set_size_inches(10, 6)
ax = plt.axes(projection='3d')
p = ax.plot_surface(x,y,z, cmap = cm.cool)
fig.colorbar(p, ax=ax)
```

```python
u = np.arange(-5, 5, 0.2)
v = np.arange(-6, 6, 0.2)

x, y = np.meshgrid(u, v)
z = np.cos(x) * np.cos(y) * np.exp(-np.sqrt(x**2 + y**2) / 20)
```

```python
fig = plt.figure()
fig.set_size_inches(10, 6)
ax = plt.axes(projection='3d')
p = ax.plot_surface(x,y,z, cmap = cm.cool)
fig.colorbar(p, ax=ax)
```
